## 1. Two Sum

```python
target = [3,5,1,3,5]
k = 6

checked = {}

def twosum(target,k):

    for i,n in enumerate(target):
        diff = k-n
        if diff in checked:
            return [checked[diff],i]
        else:
            checked[n]=i
```

```python
twosum(target,k)
```
